```python
import numpy as np

# Data for gRPC
grpc_data = {
    500: [1.453, 1.8679, 2.30028, 2.4527, 2.89693, 3.0850],
    300: [1.37490, 1.62933, 1.80063, 1.9523, 2.20973, 2.36648],
    100: [1.2170, 1.3138, 1.39299, 1.44211, 1.56853, 1.5908],
    50: [1.2224, 1.2439, 1.3194, 1.34311, 1.3899, 1.46650]
}

# Data for ZeroMQ
zeromq_data = {
    500: [1.01672, 1.602193, 1.80311, 2.00505, 2.2063, 2.4038],
    300: [0.7096, 0.9098, 1.2068, 1.37120, 1.413291, 1.70928],
    100: [0.63791, 0.706, 0.7837, 0.8334, 0.9038, 0.9231],
    50: [0.2023, 0.3015, 0.4019, 0.5020, 0.601, 0.7021]
}

# Calculate variances and standard deviations
grpc_variances = {count: np.var(times) for count, times in grpc_data.items()}
grpc_std_devs = {count: np.std(times) for count, times in grpc_data.items()}

zeromq_variances = {count: np.var(times) for count, times in zeromq_data.items()}
zeromq_std_devs = {count: np.std(times) for count, times in zeromq_data.items()}

# Print results
print("gRPC Variances:", grpc_variances)
print("gRPC Standard Deviations:", grpc_std_devs)

print("ZeroMQ Variances:", zeromq_variances)
print("ZeroMQ Standard Deviations:", zeromq_std_devs)
```
